데이터 준비하기

```python
# 최고 기온 데이터만 리스트에 담아서 출력하기

import csv

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
header = next(data) # 헤더

result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data :
  if row[-1] != "" :
    result.append(float(row[-1])) # result = [float(row[-1]) for row in data if row[-1] != ""]

print(result) # 최고 기온 리스트
```

## 날짜 데이터 추출하기

```python
date = '2021-06-14'
print(date.split('-'))
```

```python
# 최고 기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화

import csv
import matplotlib.pyplot as plt

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
header = next(data) # 헤더

result = [] # 최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data :
  if row[-1] != "" : # 결측치 제외하기
    if row[0].split('-')[1] == "08" : # 날짜를 년, 월, 일로 잘라서 월이 08 데이터 추출
      result.append(float(row[-1]))

# 그래프 크기 조절
plt.figure(dpi=200) # dpi는 해상도
plt.plot(result, "hotpink")
plt.show()
```

매년 생일 기준 최고 기온, 최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
next(data) # 헤더

high = [] # 최고 기온 리스트
low = [] # 최저 기온 리스트

for row in data :
  if row[-1] != '' and row[-2] != '' : # 최고기온과 최저기온 둘 다 결측치가 없으면
    if row[0].split('-')[1] == "02" and row[0].split('-')[2] == "29" :
      high.append(float(row[-1])) # 최고 기온을 리스트에 담기
      low.append(float(row[-2])) # 최저 기온을 리스트에 담기

# 데이터 시각화
plt.figure(dpi=200)
plt.plot(high, "r", label="maxTemp")
plt.plot(low, "b", label="minTemp")
plt.legend() # 범례 표시
plt.show()
```

날짜 구간을 좁혀서 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
next(data) # 헤더

high = [] # 최고 기온 리스트
low = [] # 최저 기온 리스트

for row in data :
  if row[-1] != '' and row[-2] != '' : # 최고기온과 최저기온 둘 다 결측치가 없으면
    if 1993 <= int(row[0].split('-')[0]) : # 문자열은 크기 비교를 할 수 없기 때문에 int()로 감싸주어서 숫자형으로 바꿔준다
      if row[0].split('-')[1] == '06' and row[0].split('-')[2] == "06" :
        high.append(float(row[-1]))
        low.append(float(row[-2]))

# 데이터 시각화
plt.figure(dpi=200)
plt.plot(high, "r", label="maxTemp")
plt.plot(low, "b", label="minTemp")
plt.legend() # 범례 표시
plt.show()
```

매년 생일 기준 최고 기온, 최저 기온, 평균 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
next(data) # 헤더

high = [] # 최고 기온 리스트
low = [] # 최저 기온 리스트
avg = []

for row in data :
  if row[-1] != '' and row[-2] != '' and row[-3] != "" : # 최고기온, 최저기온, 평균기온의 결측치 제거
    if 1993 <= int(row[0].split('-')[0]) : # 문자열은 크기 비교를 할 수 없기 때문에 int()로 감싸주어서 숫자형으로 바꿔준 후 크기 비교
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == "24" :
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        avg.append(float(row[-3]))

# 데이터 시각화
# 차트 설정
plt.figure(dpi=200) # 해상도 조절
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정 -> 한글 깨짐
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# plot 차트 그리기
plt.plot(high, "r", label="maxTemp") # 최고 기온
plt.plot(avg, "g", label="avgTemp") # 평균 기온
plt.plot(low, "b", label="minTemp") # 최저 기온

plt.title("내 생일의 기온 변화 그래프")
plt.legend() # 범례 표시
plt.show()
```
