```python
import matplotlib.pyplot as plt
import csv

# 서울특별시, 부산광역시, 대구광역시, 인천광역시, 광주광역시, 대전광역시, 울산광역시,
# 경기도, 강원도, 충청북도, 충청남도, 전라북도, 전라남도, 경상북도, 경상남도,제주특별자치도

name = '부산광역시' # 위의 광역자치단체 중 하나의 이름을 넣으면 된다.

f = open('/content/worldCup_2002.csv', encoding= 'cp949')
data2002 = csv.reader(f)

region2002 = []

for row in data2002:
  if name in row[1]:
    for i in row[4:] :
      region2002.append(float(i.replace(',','')))

f = open('/content/worldCup_2003.csv', encoding= 'cp949')
data2003 =csv.reader(f)

region2003 = []

for row in data2003:
  if name in row[1]:
    for i in row[4:] :
      region2003.append(float(i.replace(',','')))

f = open('/content/worldCup_2004.csv', encoding= 'cp949')
data2004 =csv.reader(f)

region2004 = []

for row in data2004:
  if name in row[1]:
    for i in row[4:] :
      region2004.append(float(i.replace(',','')))

plt.figure(figsize=(10, 3), dpi = 200)
plt.style.use('ggplot') # 스타일 지정
plt.plot(region2002,'r', label = '2002')
plt.plot(region2003,'b', label = '2003')
plt.plot(region2004, 'purple', label = '2004')
plt.title('02-04 %s의 인구 그래프' %name)
plt.legend()
plt.show() 
```

<strong>합계출산율</strong>

```python
import matplotlib.pyplot as plt
import csv

f = open('/content/born.csv', encoding='cp949')
data = csv.reader(f)

bornList = []
year = []

for row in data :
  x_num = 2000
  if '합계출산율' in row[0] :
    for idx in row[1:] :
      bornList.append(float(idx))
      year.append(x_num)
      x_num+=1

plt.figure(dpi = 130)
plt.style.use('bmh')
plt.plot(bornList, 'r', marker = 'o')
plt.title('2000-2016 합계출산율')
plt.show() 
```
