```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

## Q07. 연령대별 사망자 수 확인

```python
age0 = titanic[(titanic['age'] >= 0) & (titanic['age'] < 10)]
age1 = titanic[(titanic['age'] >= 10) & (titanic['age'] < 20)]
age2 = titanic[(titanic['age'] >= 20) & (titanic['age'] < 30)]
age3 = titanic[(titanic['age'] >= 30) & (titanic['age'] < 40)]
age4 = titanic[(titanic['age'] >= 40) & (titanic['age'] < 50)]
age5 = titanic[(titanic['age'] >= 50) & (titanic['age'] < 60)]
age6 = titanic[(titanic['age'] >= 60) & (titanic['age'] < 70)]
age7 = titanic[(titanic['age'] >= 70) & (titanic['age'] < 80)]
age8 = titanic[titanic['age'] >= 80]
```

```python
print("0~9세 사망자수 : %d" %len(age0[age0['survived'] == 0]))
print("10대 사망자수  : %d" %len(age1[age1['survived'] == 0]))
print("20대 사망자수  : %d" %len(age2[age2['survived'] == 0]))
print("30대 사망자수  : %d" %len(age3[age3['survived'] == 0]))
print("40대 사망자수  : %d" %len(age4[age4['survived'] == 0]))
print("50대 사망자수  : %d" %len(age5[age5['survived'] == 0]))
print("60대 사망자수  : %d" %len(age6[age6['survived'] == 0]))
print("70대 사망자수  : %d" %len(age7[age7['survived'] == 0]))
print("80세 이상 사망자수 : %d" %len(age8[age8['survived'] == 0]))
```
