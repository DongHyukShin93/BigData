```python
import pandas as pd
```

```python
df = pd.read_csv("./../../DoItPandas/data/gapminder.tsv",sep="\t")
```

```python
multi_group_var = df.groupby(["year","continent"])["lifeExp","gdpPercap"]
print(multi_group_var.mean().head(7),"\n")
```

```python
uniqueList1 = df["year"].unique()
uniqueList2 = df["continent"].unique()
for i in uniqueList1 :
    yearList = df[df["year"] == i]
    for j in uniqueList2 :
        continentList = yearList[df["continent"] == j]
        print(f"{i}년 {j:^9}지역의 기대수명 : {continentList['lifeExp'].mean():.2f}\n\t\t     1인당 GDP : {continentList['gdpPercap'].mean():.2f}")
    print() 
```
