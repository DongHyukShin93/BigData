예제에 사용할 데이터셋

1. 이진 분류 데이터셋(forge) 확인하기

데이터셋 다운로드

```python
import mglearn
import matplotlib.pyplot as plt

# 데이터셋 만들기 # 2차원 배열이라 X가 대문자
X, y = mglearn.datasets.make_forge()
```

데이터 확인하기

```python
print('X.shape :', X.shape)
print('y.shape :', y.shape)
```

```python
X # 2차원 배열
```

```python
y # 0과 1로 되어있으니 이진 분류가 가능하겠다라고 생각
```

```python
plt.figure(dpi=150)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 산점도 그리기
mglearn.discrete_scatter(X[:,0], X[:,1], y)

plt.legend(['클래스 0', '클래스 1'], loc=4)
plt.xlabel('첫 번째 특성')
plt.ylabel('두 번째 특성')
plt.show()
```

2. 회귀 데이터셋(wave) 확인하기

데이터 다운로드

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
```

데이터 확인하기

```python
print('X.shape :', X.shape) # 데이터셋은 데이터 포인트 40개와 특성 1개
print('y.shape :', y.shape) # 레이블(label, y, 정답) 연속적인 실수 형태
```

```python
X
```

```python
y
```

```python
# 산점도 그리기
# 차트 설정
plt.figure(dpi=150)
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 차트 그리기
plt.plot(X, y, 'o')
# plt.ylim(-3, 3)
plt.xlabel('특성')
plt.ylabel('타깃')
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
print(cancer.keys()) # 데이터셋 키 확인
```

데이터 확인하기

```python
cancer.data.shape # 유방암 데이터의 형태 -> 569건, 특성 30개
```

```python
# 클래스별 샘플 갯수

import numpy as np

print('클래스별 샘플 갯수 :\n', {n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))})
```
