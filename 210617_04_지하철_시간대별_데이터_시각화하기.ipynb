아침 7시 승차 데이터 추출 -> 10번 인덱스 데이터 값 추출

```python
import csv

f = open('/content/subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

result = [] # 아침 7시 승차 데이터 추출 # 10번째 인덱스에 존재

for row in data :
  row[4:] = map(int, row[4:]) # 문자열을 정수형으로 변환 후 다시 저장
  result.append(row[10])

print(len(result)) # 역의 갯수
print(result)
```

오름차순으로 데이터 정렬하여 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

result = [] # 아침 7시 승차 데이터 추출 # 10번째 인덱스에 존재

for row in data :
  row[4:] = map(int, row[4:]) # 문자열을 정수형으로 변환 후 다시 저장
  result.append(row[10])

result.sort() # 데이터를 오름차순으로 정렬

# 데이터 시각화
plt.figure(dpi=120)
plt.style.use('bmh')
plt.bar(range(len(result)), result)
plt.show()
```

출근 시간 (7-9시) **막대 그래프** 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

result = [] # 출근 시간 (7-9시) 승차 데이터 추출 # 10, 12, 14번째 인덱스에 존재

for row in data :
  row[4:] = map(int, row[4:]) # 문자열을 정수형으로 변환 후 다시 저장
  result.append(sum(row[10:13:2])) # 10,12,14 추출

result.sort() # 데이터를 오름차순으로 정렬

# 데이터 시각화
plt.figure(dpi=150)
plt.style.use('bmh')
plt.bar(range(len(result)), result)
plt.show()
```

출근 시간대에 사람들이 가장 많이 타는 역 찾기

```python
import csv

f = open('/content/subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

max = 0
max_station = ''

for row in data :
  row[4:] = map(int, row[4:]) # 문자열을 정수형으로 변환 후 다시 저장
  if sum(row[10:13:2]) > max : # 10,12 추출
    max = sum(row[10:13:2])
    max_station = row[3] + '(' + row[1] + ')'

print(max_station, max)
```
