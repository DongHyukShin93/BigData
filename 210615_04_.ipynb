1) 성별 데이터 수평 막대 그래프로 시각화 (항아리 그래프) -> 우리 동네 데이터

```python
# 라이브러리 불러오기
import csv

# 파일 열기
f = open("/content/gender.csv", encoding="cp949")
data = csv.reader(f)
# next(data) # 헤더 넘기기

male = []
female = []

for row in data :
  if '영덕동' in row[0] :
    for i in row[3:104] : # 남성 데이터 : 3 ~ 103번 데이터 슬라이싱 -> 정수형으로 변환 -> 리스트에 저장
      male.append(-int(i)) # 남성 데이터 음수로 변경
    for j in row[106:] : # 여성 데이터 : 106 ~ 끝까지 데이터 슬라이싱 -> 정수형으로 변환
      female.append(int(j))

import matplotlib.pyplot as plt
# 차트 설정
plt.figure(dpi=200) # 해상도 조절
plt.style.use("bmh")
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정 -> 한글 깨짐 처리
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.barh(range(101), male, label='남성') # 막대그래프를 수직 방향이 아닌 수평으로 그리기
plt.barh(range(101), female, label='여성') # 막대그래프를 수직 방향이 아닌 수평으로 그리기
plt.title('용인시 영덕동의 남녀 성별 인구 분포')
plt.legend()
plt.show()
```

2) 2002년 연령별 인구 구조 -> 2003년 연령별 인구 구조 비교 (같이 그리기)

```python
import csv

# 2002
f = open("/content/worldCup2002.csv", encoding='cp949')
data = csv.reader(f)
next(data)

mapo2002 = []

for row in data :
  if row[1] == "마포구" :
    for i in row[4:] :
      mapo2002.append(float(i))

# 2003
f = open("/content/worldCup2003.csv", encoding='cp949')
data2 = csv.reader(f)
next(data2)

mapo2003 = []

for row in data2 :
  if row[1] == "마포구" :
    for i in row[4:] :
      mapo2003.append(float(i))

# 2004
f = open("/content/worldCup2004.csv", encoding='cp949')
data3 = csv.reader(f)
next(data3)

mapo2004 = []

for row in data3 :
  if row[1] == "마포구" :
    for i in row[4:] :
      mapo2004.append(float(i))
```

```python
import matplotlib.pyplot as plt

# 차트 설정
plt.figure(dpi=200)
plt.style.use('bmh') # 격자 무늬 스타일로 지정
# 차트 그리기
plt.plot(mapo2002, color="r", label="2002년")
plt.plot(mapo2003, color="b", label="2003년")
plt.plot(mapo2004, color="g", label="2004년")
plt.title("02-04 마포구의 인구 구조")
plt.legend()
plt.show()
```
