꺾은선 그래프로 표현하기

```python
# 라이브러리 불러오기
import csv
import matplotlib.pyplot as plt

# 파일 열기
f = open("/content/gender.csv", encoding="cp949")
data = csv.reader(f)
# next(data) # 헤더 넘기기

# 지역 명을 입력받는 변수 선언
name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ")

# 남자와 여자의 데이터 담을 리스트 생성
m = []
f = []

for row in data :
  if name in row[0] :
    for i in row[3:104] : # 남성 데이터 : 3 ~ 103번 데이터 슬라이싱 -> 정수형으로 변환 -> 리스트에 저장
      m.append(int(i.replace(',','')))
    for j in row[106:] : # 여성 데이터 : 106 ~ 끝까지 데이터 슬라이싱 -> 정수형으로 변환
      f.append(int(j.replace(',','')))
    break # 

# 차트 설정
plt.figure(dpi=200) # 해상도 조절
plt.style.use('ggplot') # 스타일 지정
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정 -> 한글 깨짐 처리

# 꺾은선 그래프로 남녀 데이터 표현하기
plt.plot(m, label="남성")
plt.plot(f, label="여성")
plt.legend()
plt.show()
```

```python
import csv

f = open("gender.csv", encoding="CP949")
data = csv.reader(f)

male = []
female = []

townName = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data:
    if townName in row[0]:
        for i in row[3:104]: # 남성 데이터 : 3~103번 데이터 슬라이싱 - 정수 데이터로 변환 - male리스트에 저장
            male.append(int(i.replace(',','')))
        for i in row[106:]: # 남성 데이터 : 106~끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - female리스트에 저장
            female.append(int(i.replace(',','')))
        break


plt.style.use('ggplot')
plt.figure(dpi= 200)

# 꺾은선 그래프로 남여 인구수 데이터 표현하기
plt.plot(male, label = "남성")
plt.plot(female, label = "여성")
plt.legend()
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
# 라이브러리 불러오기
import csv
import matplotlib.pyplot as plt

# 파일 열기
f = open("/content/gender.csv", encoding="cp949")
data = csv.reader(f)
# next(data) # 헤더 넘기기

# 지역 명을 입력받는 변수 선언
name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ")

# 남자 인구 수 빼기 여자 인구수의 데이터 담을 리스트 생성
result = []

for row in data :
  if name in row[0] :
    for i in range(3,104) :
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',',''))) # 콤마를 공백으로 바꾼후 정수형으로 변환 후 (남성 인구 수 - 여자 인구 수)를 result리스트에 담기
    break

# 차트 설정
plt.figure(dpi=200) # 해상도 조절
plt.style.use('ggplot') # 스타일 지정
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정 -> 한글 깨짐 처리
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 차트 그리기
plt.bar(range(101), result)
plt.title(name+' 지역의 남녀 성별 인구 차이')
plt.show()
```

위치, 크기가 서로 다른 100개의 점 만들기 - random 이용

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(100) :
  x.append(random.randint(50,100))
  y.append(random.randint(50,100))
  size.append(random.randint(10,100))

plt.figure(dpi=200)
plt.style.use("ggplot")
plt.scatter(x, y, s = size)
plt.show()
```

```python
plt.figure(dpi=200)
plt.style.use("ggplot")
plt.scatter(x, y, s = size, c = size, cmap='jet')
plt.colorbar() # 차트 오른쪽에 색상 막대를 생성
plt.show()
```

```python
plt.figure(dpi=200)
plt.style.use("ggplot")
plt.scatter(x, y, s = size, c = size, cmap='jet', alpha=0.7) # alpha 0에 가까울수록 투명, 1에 가까우면 불무명, 보통 0.7~0.8을 많이 쓴다
plt.colorbar() # 차트 오른쪽에 색상 막대를 생성
plt.show()
```
