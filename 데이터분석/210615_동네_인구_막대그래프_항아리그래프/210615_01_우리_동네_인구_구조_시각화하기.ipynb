연령 별 인구 수 (0~100세 이상) 데이터 확인 -> Q 4.

```python
# 라이브러리 불러오기
import csv

# 파일 열기
f = open("/content/age.csv", encoding="cp949")
data = csv.reader(f)
# next(data) # 헤더 넘기기

for row in data :
  if '신도림' in row[0] :
    print(row[3:]) # 0세 ~ 100세 이상 인구 수 출력 # row[0]: 지역명, row[1]: 총인구수, row[2]: 연령구간인구수, row[3]~  인덱스 부터 연령별 인구 수
    print(len(row[3:])) # 0세 ~ 100세 이상 데이터 수 출력 -> 101개 출력
```

우리 동네 연령 별 인구 수 result 리스트에 저장하기 -> 시각화를 위해

```python
# 라이브러리 불러오기
import csv

# 파일 열기
f = open("/content/age.csv", encoding="cp949")
data = csv.reader(f)
# next(data) # 헤더 넘기기

result = []

for row in data :
  if '신도림' in row[0] :
    for i in row[3:] :
      result.append(int(i)) # int() 정수형으로 바꾼 후에 result 리스트에 담기
print(result)
```

우리 동네 인구 동네 데이터 시각화하기

```python
import matplotlib.pyplot as plt

# 차트 설정
plt.figure(dpi=200)
plt.style.use('bmh') # 격자 무늬 스타일로 지정
# 차트 그리기
plt.plot(result)
plt.show()
```
