지역명을 입력받아서 연령 별 인구 수 데이터 시각화하기

```python
# 라이브러리 불러오기
import csv
import matplotlib.pyplot as plt

# 파일 열기
f = open("/content/age.csv", encoding="cp949")
data = csv.reader(f)
# next(data) # 헤더 넘기기

# 지역 명을 입력받는 변수 선언
name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ")

result = []

for row in data :
  if name in row[0] :
    for i in row[3:] :
      result.append(int(i.replace(',',''))) # int() 정수형으로 바꾼 후에 result 리스트에 담기 # 만약 인구 수 가 1000을 넘으면 1,000 이런 형태로 표현이 되어서 i.replace(',','')를 통해 ,(콤마)를 제거해야 한다

# 차트 설정
plt.figure(figsize=(10,3), dpi=200) # 해상도 조절
plt.style.use('bmh') # 스타일 지정
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정 -> 한글 깨짐 처리
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
# plot 차트 그리기
plt.plot(result)
# plt.plot(maxAgeList, ,'k.',label="maxAge는 959명")
# plt.plot(minAgeList, ,'r.',label="minAge는 2명")
plt.title(name+'의 인구 구조')
# plt.legend()
plt.show()
```

성별 데이터 저장하기(Type1)

```python
# 라이브러리 불러오기
import csv
import matplotlib.pyplot as plt

# 파일 열기
f = open("/content/gender.csv", encoding="cp949")
data = csv.reader(f)

# 남자와 여자의 데이터 담을 리스트 생성
m = []
f = []

for row in data :
  if '신도림' in row[0] :
    for i in range(101) :
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)])) # -1 값부터 담음 -> 100세 이상 부터 값이 담김
print(f)
f.reverse() # 0세 부터 인구 수가 나올 수 있도록 데이터 뒤집기
print(f)
```

성별 데이터 저장하기(Type2)

```python
# 라이브러리 불러오기
import csv
import matplotlib.pyplot as plt

# 파일 열기
f = open("/content/gender.csv", encoding="cp949")
data = csv.reader(f)

# 남자와 여자의 데이터 담을 리스트 생성
m = []
f = []

for row in data :
  if '신도림' in row[0] :
    for i in row[3:104] : # 남성 데이터 : 3 ~ 103번 데이터 슬라이싱 -> 정수형으로 변환 -> 리스트에 저장
      m.append(int(i))
    for j in row[106:] : # 여성 데이터 : 106 ~ 끝까지 데이터 슬라이싱 -> 정수형으로 변환
      f.append(int(j))
print(m)
print(f)
```

```python
import matplotlib.pyplot as plt

# 차트 설정
plt.figure(figsize=(10,3), dpi=200) # 해상도 조절
plt.style.use('bmh') # 스타일 지정
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정 -> 한글 깨짐 처리
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.figure(dpi=100)
plt.barh(range(101), m, label="남성") # 막대그래프를 수직 방향이 아닌 수평으로 그리기
plt.barh(range(101), f, label="여성") # 막대그래프를 수직 방향이 아닌 수평으로 그리기
# plt.title('용인시 기흥구 영덕동 인구 구조')
plt.legend()
plt.show()
```

항아리 그래프를 위해 남성 데이터 음수로 바꾸기

```python
# 라이브러리 불러오기
import csv

# 파일 열기
f = open("/content/gender.csv", encoding="cp949")
data = csv.reader(f)

# 남자와 여자의 데이터 담을 리스트 생성
m = []
f = []

for row in data :
  if '신도림' in row[0] :
    for i in row[3:104] : # 남성 데이터 : 3 ~ 103번 데이터 슬라이싱 -> 정수형으로 변환 -> 리스트에 저장
      m.append(-int(i)) # 남성 데이터 음수로 변경
    for j in row[106:] : # 여성 데이터 : 106 ~ 끝까지 데이터 슬라이싱 -> 정수형으로 변환
      f.append(int(j))
```

```python
# 차트 설정
plt.figure(dpi=200) # 해상도 조절
# plt.style.use('bmh') # 스타일 지정
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정 -> 한글 깨짐 처리
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.figure(dpi=150)
plt.barh(range(101), m, label='남성') # 막대그래프를 수직 방향이 아닌 수평으로 그리기
plt.barh(range(101), f, label='여성') # 막대그래프를 수직 방향이 아닌 수평으로 그리기
plt.title('신도림의 남녀 성별 인구 분포')
plt.legend()
plt.show()
```
