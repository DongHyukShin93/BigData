데이터 가져오기

```python
df = pd.read_csv('/content/TASHU.csv')
df.info()
```

지도에 Markers를 이용한 타슈 위치 표시하기

```python
# 대전 중심으로 map 열기
map = folium.Map(location=[36.351,127.386], zoom_start=13)

# 타슈의 상태에 따라 Marker를 다른 색으로 표시하기
datail_location = df['위치']
condition = df.상태정보

for index_draw in range(1, len(condition)) :

  if condition[index_draw] == '정상' :
    # 정상인 경우 파란색
    folium.Marker([df.loc[index_draw, '위도'], df.loc[index_draw, '경도']],
                   popup='정상', icon=folium.Icon(color='blue', icon='info-sign')).add_to(map)
  elif condition[index_draw] == 'Network 에러' :
    # 에러인 경우 빨간색
    folium.Marker([df.loc[index_draw, '위도'], df.loc[index_draw, '경도']],
                   popup='Network 에러', icon=folium.Icon(color='red', icon='info-sign')).add_to(map)
  else :
    # 그 외(학생들집)은 초록색으로
    folium.Marker([df.loc[index_draw, '위도'], df.loc[index_draw, '경도']],
                   popup='집', icon=folium.Icon(color='green', icon='info-sign')).add_to(map)
  folium.Marker([36.37347861756061, 127.36010849501348],
                   popup='KAIST', icon=folium.Icon(color='green', icon='star')).add_to(map)
```
