최고 기온 데이터 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
next(data) # 헤더

result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data :
  if row[-1] != "" :
    result.append(float(row[-1]))

# result = [float(row[-1]) for row in data if row[-1] != ""]

# 최고 기온 데이터 히스토그램
plt.figure(dpi=150)
plt.hist(result, bins=100, color="r")
plt.show()
```

1월과 8월의 데이터를 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
next(data) # 헤더

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = [] # 1월의 최고 기온을 담을 jun 리스트 변수 선언

for row in data :
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터 추출
  if row[-1] != "" :
    if month == "08" : # 8월 데이터만
      aug.append(float(row[-1])) # aug 변수에 담기
    if month == "01" : # 1월 데이터만
      jan.append(float(row[-1])) # jun 변수에 담기

# 히스토그램 그리기
plt.figure(dpi=150)
plt.hist(aug, bins=100, color="r", label="Aug")
plt.hist(jan, bins=100, color="b", label="Jan")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
next(data) # 헤더

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = [] # 1월의 최고 기온을 담을 jan 리스트 변수 선언
apr = [] # 6월의 최고 기온을 담을 apr 리스트 변수 선언

for row in data :
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터 추출
  if row[-1] != "" :
    if month == "08" : # 8월 데이터만
      aug.append(float(row[-1])) # aug 변수에 담기
    if month == "01" : # 1월 데이터만
      jan.append(float(row[-1])) # jan 변수에 담기
    if month == "04" : # 6월 데이터만
      apr.append(float(row[-1])) # apr 변수에 담기

# 히스토그램 그리기
plt.figure(dpi=150)
plt.hist(aug, bins=100, color="r", label="Aug")
plt.hist(jan, bins=100, color="b", label="Jan")
plt.hist(apr, bins=100, color="g", label="April")
plt.legend()
plt.show()
```

1월부터 12월까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
next(data) # 헤더

month = [[],[],[],[],[],[],[],[],[],[],[],[]] # 월별 데이터 저장할 month 변수 선언

for row in data :
  if row[-1] != "" :
    month[int(row[0].split('-')[1])-1].append(float(row[-1])) # 월과 같은 번호의 인덱스에 월별 데이터 저장 (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1]

plt.figure(dpi=150)
plt.boxplot(month)
plt.show()
```

8월 1일부터 31일까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/seoul.csv", encoding="cp949")
data = csv.reader(f)
next(data) # 헤더

day = [] # 일별 데이터를 저장할 리스트 day 변수 선언

for i in range(31) :
  day.append([]) # day 리스트 내에 31개 리스트 생성 [[],[],[],...31개]

for row in data :
  if row[-1] != "" :
    if row[0].split('-')[1] == "08" :
      day[int(row[0].split('-')[2])-1].append(float(row[-1])) # 8월 최고 기온 값 저장 day[0] -> 1일

plt.figure(figsize=(10,5), dpi=150)
plt.boxplot(day, showfliers=False) # 아웃라이어(이상치) 값 생략
plt.show()
```
